```python
%%sql
select
  s.orderdate,
  s.quantity*s.netprice*s.exchangerate as net_revenue,
  p.productkey,
  p.categoryname,
  p.subcategoryname,
  p.productkey,
  c.customerkey,
  c.countryfull,
  c.surname,
  c.givenname,
case when s.quantity*s.netprice*s.exchangerate > 1000 then 'High' else 'low' end as high_low
from
  sales s
left join product p on s.productkey = p.productkey
left join customer c on s.customerkey = c.customerkey
where orderdate::date >= '2020-01-01'
```

```python
%%sql

select
s.orderdate,
count(distinct s.customerkey),
count(distinct case when c.continent = 'Europe' then s.customerkey end) as eu_customers,
count(distinct case when c.continent = 'North America' then s.customerkey end) as na_customers,
count(distinct case when c.continent = 'Australia' then s.customerkey end) as au_customers
from sales s
left join customer c on s.customerkey = c.customerkey
where s.orderdate between '2023-01-01' and '2023-12-31'
group by s.orderdate
order by s.orderdate
```

```python
%%sql

select
p.categoryname,
PERCENTILE_CONT(0.5) within group  (order by (case when s.orderdate between '2022-01-01' and '2022-12-31' then s.quantity*s.netprice*s.exchangerate end)) as y2022_median,
PERCENTILE_CONT(0.5) within group  (order by (case when s.orderdate between '2023-01-01' and '2023-12-31' then s.quantity*s.netprice*s.exchangerate end)) as y2023_median

from sales s
left join product p on s.productkey = p.productkey
group by p.categoryname
```

```python
%%sql
select
  PERCENTILE_CONT(0.50) within group (order by (s.quantity*s.netprice*s.exchangerate)) as median
from sales s
left join product p on p.productkey = s.productkey
```

```python
%%sql
with cte as (
select
  PERCENTILE_CONT(0.50) within group (order by (s.quantity*s.netprice*s.exchangerate)) as median
from sales s
where s.orderdate between '2022-01-01' and '2023-12-31'
)


select
  p.categoryname,
  sum(s.quantity*s.netprice*s.exchangerate) as net_revenue,
  SUM(case when (s.quantity*s.netprice*s.exchangerate) < (select median from cte)
    and (s.orderdate between '2022-01-01' and '2022-12-31')
    then (s.quantity*s.netprice*s.exchangerate) end ) as low_net_revenue_2022,
  SUM(case when (s.quantity*s.netprice*s.exchangerate) >= (select median from cte)
    and (s.orderdate between '2022-01-01' and '2022-12-31')
    then (s.quantity*s.netprice*s.exchangerate) end ) as high_net_revenue_2022
from sales s
left join product p on p.productkey = s.productkey
group by p.categoryname
```

```python
%%sql

SELECT
    PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY (s.quantity * s.netprice * s.exchangerate)) AS median
FROM
    sales s
WHERE
    orderdate BETWEEN '2022-01-01' AND '2023-12-31'
```

```python
%%sql
select
  PERCENTILE_CONT(0.50) within group (order by (s.quantity*s.netprice*s.exchangerate)) as median
from sales s
```

```python
%%sql
with cte as (
  select
    PERCENTILE_CONT(0.25) within group(order by (s.quantity*s.netprice*s.exchangerate)) as low_percentile,
    PERCENTILE_CONT(0.75) within group(order by (s.quantity*s.netprice*s.exchangerate)) as high_percentile
  from sales s
  where s.orderdate between '2022-01-01' AND '2023-12-31'
)

select
p.categoryname,
sum(s.quantity*s.netprice*s.exchangerate) as total_revenue,
case
  when (s.quantity*s.netprice*s.exchangerate) <= cte.low_percentile then '3-low'
  when (s.quantity*s.netprice*s.exchangerate) >= cte.high_percentile then  '1-high'
  else '2-medium' end as teir
from sales s
left join product p on p.productkey = s.productkey, cte

group by p.categoryname, teir
order by p.categoryname, teir
```
